```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv("Crime_Data_from_2020_to_Present.csv")
df.head(5)
```

```python
df.isnull().sum()
```

```python
sns.heatmap(df.isnull(),yticklabels=False)
```

```python
df.describe()
```

```python
df.fillna(0)
```

```python
#Top Five Crimes
fig,ax = plt.subplots(figsize=(5,5))
grp_index = df.groupby('Crm Cd Desc')['Crm Cd Desc'].value_counts().sort_values(ascending=False).index[:5]
grp_count = [df.groupby('Crm Cd Desc')['Crm Cd Desc'].value_counts().sort_values(ascending=False)[:5][i] for i in range(5)]
ax.barh(grp_index,grp_count)
ax.set_xlabel("Number of people attempted crime")
ax.set_ylabel("Crime Label")
plt.yticks(rotation=45)
plt.show()
```

```python
#Top 5 Crime Prone Area
fig,ax = plt.subplots(figsize=(5,5))
area_crm_count = [df.groupby('AREA NAME')['AREA NAME'].value_counts().sort_values(ascending= False)[i] for i in range(5)]
area_crm_index = df.groupby('AREA NAME')['AREA NAME'].value_counts().sort_values(ascending= False).index[:5]
ax.bar(area_crm_index,area_crm_count,color="teal")
```
